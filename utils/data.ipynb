```python
import matplotlib.pyplot as plt
import pandas as pd

import config
```

```python
def load_csv_data(path: str) -> pd.core.frame.DataFrame:
    """Load CSV dataset as dataframe"""
    return pd.read_csv(path, delimiter=',')
```

```python
def plot_test_data(input_data: any) -> None:
    plt.figure(figsize=(10, 6))
    
    y_min = input_data.min()
    y_max = input_data.max()

    plt.plot(input_data, color='blue', label='Input Data')
    plt.xlabel("Time Step t")
    plt.ylabel('Electricity load [MW]')
    plt.legend(bbox_to_anchor=(1.02, 1.00), loc='upper left', borderaxespad=0, fontsize=10)
    plt.show()
```

```python
df = load_csv_data(config.test_data_path)
df
```

```python
input_data = df["input"]
input_data
```

```python
plot_test_data(input_data)
```

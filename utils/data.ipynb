```python
from typing import Tuple

import matplotlib.pyplot as plt
import pandas as pd

import config
```

```python
def load_csv_data(path: str) -> pd.core.frame.DataFrame:
    """Load CSV dataset as dataframe"""
    return pd.read_csv(path, delimiter=',')
```

```python
def get_min_max(df: pd.core.frame.DataFrame, start_index: int, end_index: int) -> Tuple[float, float]:
    subset = df.iloc[start_index:end_index]
    max_value = subset.max()
    min_value = subset.min()

    return min_value, max_value
```

```python
def plot_test_data(input_data: any) -> None:
    plt.figure(figsize=(10, 6))

    y_min = input_data.min()
    y_max = input_data.max()

    plt.plot(input_data, color='blue', label='Input Data')
    plt.axhline(y_min, linestyle="--", color='black', label='Min')
    plt.axhline(y_max, linestyle="--", color='black', label='Max')
    plt.annotate(f'{y_min}', xy=(0, y_min), xytext=(-25000, y_min+110), ha='right', va='top', color='black', arrowprops=dict(arrowstyle='-', color='black'))
    plt.annotate(f'{y_max}', xy=(0, y_max), xytext=(-25000, y_max-110), ha='right', va='bottom', color='black', arrowprops=dict(arrowstyle='-', color='black'))

    plt.xlabel("Time Step t")
    plt.ylabel('Electricity load [MW]')
    # plt.legend(bbox_to_anchor=(1.02, 1.00), loc='upper left', borderaxespad=0, fontsize=10)
    plt.show()
```

```python
df = load_csv_data(config.test_data_path)
df
```

```python
input_data = df["input"]
input_data
```

```python
plot_test_data(input_data)
```
